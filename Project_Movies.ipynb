### Data Cleaning

We will begin data cleaning by checking for null values.

```python
#Check for null values in the dataset
df.isnull().sum()
```

From the list of null values we can see two columns that contain a large amount of null values.
The 'homepage' column has 3091 null values and the 'tagline' columns contains 844 null values. 
We are going to drop those two colums from the dataframe as they don't contain any useful information for the purposes of this analysis.

```python
#Drop the columns with the most null values
df.drop(columns=['homepage','tagline'], inplace=True)
```

```python
#Check the remaining null values
df.isnull().sum()
```

We can see now that the two columns were dropped successfully. Now there are only a few remaining null values left to deal with.
We want to remove the null values that exist in the 'runtime' and 'release date' columns. There are potentially only 6 rows to drop from the remianing 4803, so they will have negligible effect on our overall dataset.

```python
# Drop rows with any null values
df.dropna(
    axis = 'rows', 
    how = 'any', # If any null values are present, drop that entire row
    subset = ['runtime', 'release_date', 'overview'],   # The columns we found to have null values
    inplace = True # Makes change permanent
)
```

```python
#Check for remaing null values
df.isnull().sum()
```

There are no nulls remaning and we can proceed with our data wrangling

We also want to check for missing data. This is data that doesn't show as null but doesn't give us any useful information. The main check here is for values that are equal to zero when they shouldn't be.

```python
# Check number of rows containing the value zero for each column
num_rows_with_zero = df.eq(0).sum() 

# Print the number of rows with 0 in each column
print("Number of rows with 0 in each column:")
print(num_rows_with_zero)
```

There are a lot of rows where with a value of 0. This missing data must be removed.

```python
#New dataframe without the missing values.
new_df = df.loc[(df['budget'] != 0) #Excluding rows where budget = 0
                & (df['revenue'] != 0) #Excluding rows where revenue = 0
                & (df['vote_average'] != 0) #Excluding rows where vote_average = 0
                & (df['vote_count'] != 0)] #Excluding rows where vote_count = 0
```

```python
# Check that the new dataframe contains no values equal to zero
num_rows_with_zero = new_df.eq(0).sum()  # Count the number of rows with 0 in each column

# Print the number of rows with 0 in each column
print("Number of rows with 0 in each column:")
print(num_rows_with_zero)
```

Now there are no missing values and no null values in our dataset, we will do one last cleaning check for duplicates.

```python
## Function to extract genres for each datapoint
import ast ## This module can be used to evaluate literals, eg: transform string-lists back into lists
def extract_genres(x): 
    x = ast.literal_eval(x) ## Transform '['name', 'id']' back into ['name', 'id']
    Genres = [] ## Empty list to store the genres
    
    ''' Iterate through each dictionary '''
    for item in x: ## iterate for each dictionary in our list
        Genres.append(item['name']) ## Grab the 'name' key for each dictionary
        \
    ''' Return the Genres'''
    return Genres

## Apply the function to genres
df['extracted_genres'] = df['genres'].apply(extract_genres)
```

#### Budget v. Revenue:

Here we will undertake some intial data exploration. We will start by looking at the budget of films and how that effects measurable KPIs such as revenue and popularity.

```python
#Create a new dataframe to explore the effects of budget
#Dataframe contains only numeric values for purposes of correlation
budget_df = new_df[['title','budget','popularity','revenue','runtime','vote_average','vote_count']]\
    .sort_values('budget', ascending=False)\
    .set_index('title')\
    .copy()
budget_df.head()
```

For our intial investigation we will explore how movie budgets effect revenue.

```python
# Create a line plot to visualize the relationship between budget and revenue
plt.figure(figsize=(12, 6))  # Set the figure size
sns.lineplot(data=budget_df, x='budget', y='revenue', color='#009DDC')  # Plot the line plot
sns.regplot(data=budget_df, x='budget', y='revenue', scatter=False, color='#F26430', line_kws={"linestyle": "--"}, ci=None)  # Add a dashed trend line without confidence intervals
plt.title('Line Graph of Budget v. Revenue')  # Set the title of the plot
plt.xlabel('Budget ($M)')  # Set the x-axis label
plt.ylabel('Revenue ($M)')  # Set the y-axis label
plt.show()  # Display the plot
```

#### Key findings:
<ul>
    <li>There is a postive, linear relationship between budget and revenue.</li> 
    <li>This means that the higher the budget of a movie, the greater the revenue it will generate. This is what we expected to see.</li>
<li>There is one movie that appears to be a outlier, where an abnormally high revenue was generated for its budget.</li>
</ul>

### 8. Two Movies Comparison
#### Avatar v. Titanic

Haivng explored the effect of budget on revenue and popularity, we will now look closer at the two films which have the highest revenues. This is so we can see what similarities/differences they have and we will see if we can draw any conclusions as to why they are so successful.

```python
# Creating a new dataframe containing only the two films we are looking at.
highest_budgets = new_df.sort_values('revenue',  
                                     ascending=False)\
                                    .head(2)\
                                    .copy()  

highest_budgets
```

For the purposes of evaluation, we want to add a new column to the dataframe representing value for money.

We will use value for money to see which of the two films returned the most reveune per amount of budget spent. This will show which one of the movies were most efficent with their spending and got the best return for its investors.

We calculate value for money as follows:

            value for money = (revenue / budget) x 100

```python
# Create a new column 'value_for_money'
highest_budgets['value_for_money'] = (highest_budgets['revenue'] / highest_budgets['budget']) * 100

#Using indexing to check the column has been created successfully.
highest_budgets.iloc[:, [5, -1]]
```

```python
dataset = highest_budgets

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # 2 rows, 2 columns

# Plot different aspects of the dataset on separate subplots

#Bar chart comparing the popularity score for both movies
sns.barplot(data=dataset, x='title', y='popularity', ax=axs[0, 0], palette=custom_palette)
axs[0, 0].set_title('Popularity')
axs[0, 0].set_xlabel('Title')
axs[0, 0].set_ylabel('Popularity')

#Bar chart showing revenue of both movies
sns.barplot(data=dataset, x='title', y='revenue', ax=axs[0, 1], palette=custom_palette)
axs[0, 1].set_title('Revenue')
axs[0, 1].set_xlabel('Title')
axs[0, 1].set_ylabel('Revenue')

#Bar chart showing vote_average of both movies
sns.barplot(data=dataset, x='title', y='budget', ax=axs[1, 0], palette=custom_palette)
axs[1, 0].set_title('Budget')
axs[1, 0].set_xlabel('Title')
axs[1, 0].set_ylabel('Budget')

#Bar chart showing value for money
sns.barplot(data=dataset, x='title', y='value_for_money', ax=axs[1, 1], palette=custom_palette)
axs[1, 1].set_title('Value for Money')
axs[1, 1].set_xlabel('Title')
axs[1, 1].set_ylabel('Value for Money')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

#### Key findings: 
<ul>
    <li>Avatar is more popualr than Titanic.</li>
    <li>Avatar genrated higher revenues than Titanic.</li>  
     <li>Avatar had a higher budget than Titanic.</li>
    <li>Avatar represents greater value for money than Titanic.</li>
    <li>From this we can conclude that Avatar is the more succesful movie across all measureable metrics availbale to us.</li>
    <li>Despite both appearing as outliers in the overall dataset, these two blockbusters still show the same patterns as the rest of the dataset; hihgher budgets result in higher populartiy and higher revneue.</li>
